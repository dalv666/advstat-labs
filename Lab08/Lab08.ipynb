```python
import numpy as np, pandas as pd
data = pd.read_csv('ds-boot-2.csv', sep='\t' )
data
```

```python
def bootstrap_resample(X, n=None):
    """ Bootstrap resample an array_like
    Parameters
    ----------
    X : array_like
      data to resample
    n : int, optional
      length of resampled array, equal to len(X) if n==None
    Results
    -------
    returns X_resamples
    """
    if isinstance(X, pd.Series):
        X = X.copy()
        X.index = range(len(X.index))
    if n == None:
        n = len(X)
        
    resample_i = np.floor(np.random.rand(n)*len(X)).astype(int)
    X_resample = np.array(X[resample_i])  # TODO: write a test demonstrating why array() is important
    return X_resample
```

```python
def means(label):
    print(label,':')
    X = data[label]

    bootstrap_means = []
    bootstrap_median = []

    for i in range(1000):
        X_resample = bootstrap_resample(X, n=100)
        bootstrap_means.append(X_resample.mean())
        bootstrap_median.append(np.median(X_resample))
        
    np_bootstrap_means = np.array(bootstrap_means)
    np_bootstrap_medians = np.array(bootstrap_means)
    
    print('original mean:', X.mean())
    print('bootstrap mean:', np_bootstrap_means.mean())
    print('95% interval: [', np_bootstrap_means.mean() - 1.96*np.std(np_bootstrap_means),',' ,np_bootstrap_means.mean() + 1.96*np.std(np_bootstrap_means),']')
    print('')
    print('original median:', np.median(X))
    print('bootstrap median:', np.mean(np_bootstrap_medians))
    print('95% interval: [', np_bootstrap_medians.mean() - 1.96*np.std(np_bootstrap_medians), ',' , np_bootstrap_medians.mean() + 1.96*np.std(np_bootstrap_medians),']')

    
labels = ['y1','y2','y3','y4','y5']
for label in labels:
    means(label)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
features = ['p1', 'p2', 'p3', 'p4', 'p5', 'p6']
for label in labels:
    model = LinearRegression()
    model.fit(data[features], data[label])
    print(label, model.coef_)
```

```python
def bootstrap_coef_resample(X, Y, B=1000):
    n = X.shape[0]
    coeffs = []
    for b in range(B):
        model = LinearRegression()
        idx = np.random.randint(n, size=n)
        model.fit(X.as_matrix()[idx, :], Y.as_matrix()[idx])
        coeffs.append(np.append(model.coef_, model.intercept_))
    coeffs = np.array(coeffs)
    conf_int = np.percentile(coeffs, [2.5, 97.5], axis=0)
    return coeffs.mean(axis=0), conf_int

for label in labels:
    coeffs, intervals = bootstrap_coef_resample(data[features], data[label])
    
    for i in range(len(labels)):
        print(label,':', coeffs[i], '[', intervals[0][i], ',', intervals[1][i],']\n')
```

```python
from scipy.optimize import minimize
import matplotlib.pyplot as plt

data = pd.read_csv('ds-boot-1.csv', sep='\t')
data
```

```python
def go_s_shaped(t, a, b):
    return a * (1 - (1 + b * t) * np.exp(-b * t))

def go_s_obj_funct(params):
    t = np.arange(len(data['cummBugs']))
    return ((data['cummBugs'] - go_s_shaped(t, params[0], params[1])) ** 2).sum() 

#optimal parameters for G-O S-shaped model
go_res = minimize(go_s_obj_funct, [5000, 1.0], method='Nelder-Mead')
print('G-O S-shaped model', go_res.x, go_res.fun, go_res.success)

x = np.arange(len(data['cummBugs']))
xs1 = go_s_shaped(x, go_res.x[0], go_res.x[1])

plt.plot(data['cummBugs'])
plt.plot(xs1)
plt.show()
```
